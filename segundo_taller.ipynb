## Parte 1
Cargue la base de datos Lluvia_full.csv, codifique la variable LluviaMan como 0 y 1 para los días sin lluvia y con lluvia respectivamente. Además, realice una separación de la base de datos en un set de entrenamiento y set de validación. Utilice una proporción de 70:30 respectivamente estratificado por la LluviaMan. Para poder replicar sus resultados, fije una semilla antes de obtener los indices. Para ello utilice la función set.seed(2023).

```python
import pandas as pd
```

```python
# Cargar base de datos lluvia_full.csv
df = pd.read_csv('data/lluvia_full.csv', sep=',')
```

```python
# Cargar head de datos
df.head()
```

```python
# Verificar si hay datos nulos
df.isnull().sum()
```

```python
# Codifique la variable LluviaMan como 0 y 1 para los días sin lluvia y con lluvia respectivamente
df["LluviaMan"].replace(["No","Yes"], [0, 1], inplace = True)
df['LluviaMan'].value_counts()
```

```python
# Información de dataframe
df.info()
```

```python
# Separar datos estratificads por la variable LluviaMan

from sklearn.model_selection import train_test_split

df_y = df["LluviaMan"]
df_x = df.copy()
df_x.drop(["LluviaMan"], axis = "columns", inplace =  True)
# df_x = pd.get_dummies(df_x, dtype = float) // Voy a probar no dumificando acá

```

```python
# Ver información de los datos erificando datos de df_x 

df_x.info()
```

```python
# Ver información de los datos 

df_y.info()
```

```python
# Verificar datos iniciales 
df_y.head()
```

```python
# Cambiar df_y a float
# df_y = df_y.astype(float) comentado porque no dummifiqué en este paso (este es el tercer intento)
```

```python
# Separo en 70% entrenamiento y 30% testeo con una semilla 2023

x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, train_size = 0.7, random_state = 2023, stratify = df_y)
```

```python
# Escalar datos numericos en set de entrenamiento
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier


se = StandardScaler()      # instanciamos StandardScaler
ohe = OneHotEncoder(sparse_output=False) # instanciamos One Hot Encoder



# Crear el objeto ColumnTransformer
preprocessor = make_column_transformer(
    (se, x_train.dtypes == np.float64),
    (ohe, x_train.dtypes == "object")
)

# Crear el pipeline completo
pipeline = make_pipeline(
    preprocessor)

# Ajustar el pipeline a los datos
pipeline.fit(x_train, y_train)
```

```python
#Definición de modelos

logistic_model = LogisticRegression()
svm_model = SVC()
tree_model = DecisionTreeClassifier()
forest_model = RandomForestClassifier()
```

```python
x_train.head()
```

```python
# Convertir a dummies las variables categóricas 'LluviaHoy', 'Koppen', 'Estacion'
x_train = pd.get_dummies(x_train, dtype = float)
```

```python
x_train.head()
```

```python
# Aprendermos del modelo
logistic_model.fit(x_train, y_train)
svm_model.fit(x_train, y_train)
tree_model.fit(x_train, y_train)
forest_model.fit(x_train, y_train)
```

```python
# Prueba de los modelos 

y_pred_logistic = logistic_model.predict(x_train)
y_pred_svm = svm_model.predict(x_train)
y_pred_tree = tree_model.predict(x_train)
y_pred_forest = forest_model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calcular métricas para Regresión Logística
accuracy_logistic = accuracy_score(y_train, y_pred_logistic)
precision_logistic = precision_score(y_train, y_pred_logistic)
recall_logistic = recall_score(y_train, y_pred_logistic)
f1_logistic = f1_score(y_train, y_pred_logistic)

# Calcular métricas para SVM
accuracy_svm = accuracy_score(y_train, y_pred_svm)
precision_svm = precision_score(y_train, y_pred_svm)
recall_svm = recall_score(y_train, y_pred_svm)
f1_svm = f1_score(y_train, y_pred_svm)

# Calcular métricas para Árbol de Clasificación
accuracy_tree = accuracy_score(y_train, y_pred_tree)
precision_tree = precision_score(y_train, y_pred_tree)
recall_tree = recall_score(y_train, y_pred_tree)
f1_tree = f1_score(y_train, y_pred_tree)

# Calcular métricas para Random Forest
accuracy_forest = accuracy_score(y_train, y_pred_forest)
precision_forest = precision_score(y_train, y_pred_forest)
recall_forest = recall_score(y_train, y_pred_forest)
f1_forest = f1_score(y_train, y_pred_forest)

# Mostrar las métricas
print(f'Regresión Logística:')
print(f'  Exactitud: {accuracy_logistic:.2f}')
print(f'  Precisión: {precision_logistic:.2f}')
print(f'  Recuperación: {recall_logistic:.2f}')
print(f'  Puntuación F1: {f1_logistic:.2f}')

print(f'SVM:')
print(f'  Exactitud: {accuracy_svm:.2f}')
print(f'  Precisión: {precision_svm:.2f}')
print(f'  Recuperación: {recall_svm:.2f}')
print(f'  Puntuación F1: {f1_svm:.2f}')

print(f'Árbol de Clasificación:')
print(f'  Exactitud: {accuracy_tree:.2f}')
print(f'  Precisión: {precision_tree:.2f}')
print(f'  Recuperación: {recall_tree:.2f}')
print(f'  Puntuación F1: {f1_tree:.2f}')

print(f'Random Forest:')
print(f'  Exactitud: {accuracy_forest:.2f}')
print(f'  Precisión: {precision_forest:.2f}')
print(f'  Recuperación: {recall_forest:.2f}')
print(f'  Puntuación F1: {f1_forest:.2f}')
```
